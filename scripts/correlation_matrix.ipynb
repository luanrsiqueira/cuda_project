```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar dataset
df = pd.read_csv('./datasets/dataset.csv', sep=',')
df = df[['compute_version','registers','smem','cmem','num_of_cores','L2','bandwith','theoretical_flops','occupancy','input_size','duration','block_x']]

# Calcular a matriz de correlao
correlation_matrix = df.corr()

# Plotar a matriz de correlao
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', annot_kws={"size": 10})
plt.title('Matriz de Correlao', fontsize=18)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Aumentar o tamanho da fonte para 12
plt.yticks(fontsize=12)  # Aumentar o tamanho da fonte para 12
plt.show()
```
