```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Carregar o dataset
df = pd.read_csv('./datasets/dataset.csv', sep=',')
df1 = df[['compute_version','registers','smem','cmem','num_of_cores','L2','bandwith','theoretical_flops','AppId','number_of_lines_kernel','occupancy','input_size','duration','block_x']]
df1 = df1.select_dtypes(exclude=['object'])

# Plotar a distribuio em frequncia da varivel "duration"
plt.figure(figsize=(12, 8))
plt.hist(df1['duration'], bins=30, edgecolor='black', color='#1f77b4', linewidth=1.5)
plt.title('Distribuio em Frequncia da Varivel "Duration"', fontsize=25)
plt.xlabel('Durao em segundos', fontsize=20)
plt.ylabel('Frequncia', fontsize=20)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
# Aplicar a transformao logartmica  varivel "duration"
df1['log_duration'] = np.log(df1['duration'])

# Plotar a distribuio em frequncia da varivel "log_duration"
plt.figure(figsize=(12, 8))
plt.hist(df1['log_duration'], bins=30, edgecolor='black', color='#1f77b4', linewidth=1.5, density=True)
sns.kdeplot(df1['log_duration'], color='black', linewidth=2)
plt.title('Distribuio em Frequncia da Varivel "log(duration)"', fontsize=25)
plt.xlabel('Log da Durao', fontsize=20)
plt.ylabel('Frequncia', fontsize=20)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
