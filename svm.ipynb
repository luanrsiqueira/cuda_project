```python
import pandas as pd
import numpy  as np
import matplotlib as plt
from sklearn import svm
```

```python
df = pd.read_csv('./Teste-30-09.csv')

df1 = df[['AppName','compute_version','registers','smem','cmem','num_of_cores','L2','bandwith','theoretical_flops','AppId','number_of_lines_kernel','occupancy','input_size','duration','block_x']]

df1 = df.sort_values(by='AppId', ascending=True)
df1 = df1.select_dtypes(exclude=['object'])
```

```python
from sklearn import metrics

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

for idApp in range(1,10):
    idApp = float(idApp)
    X_train = df1[df1['AppId'] != idApp]
    Y_train = df1[df1['AppId'] != idApp]['duration']
    X_test = df1[df1['AppId'] == idApp]
    Y_test = df1[df1['AppId'] == idApp]['duration']

    #yTrain = np.log(Y_train + 0.01)
    #yTest = np.log(Y_test + 0.01)
    #xTrain = np.log(X_train + 0.000001)
    #xTest = np.log(X_test + 0.000001)

    labels_enc = LabelEncoder()
    y_train_labels_fit = labels_enc.fit(Y_train)
    y_train_lables_trf = labels_enc.transform(Y_train)


    modelo = LinearRegression()
    modelo.fit(X_train, y_train_lables_trf)
    #acuracia = modelo.score(X_train, yTrain)
    #print("Application: ", idApp, "Acuracia = {:0.2f}".format(acuracia))

    y_pred = modelo.predict(X_test)
    #print(regressor.score())

    #df2=pd.DataFrame({'Actual':yTest, 'Predicted':y_pred})
    print('Application: ', idApp, ' Percentage:', mean_absolute_percentage_error(Y_test, labels_enc.inverse_transform(y_pred)))
```
